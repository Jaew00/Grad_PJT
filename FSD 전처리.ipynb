## train_data 전처리

```python
raw_df = pd.read_csv('/Users/anjaeu/Desktop/UrbanSound8K/FSDKaggle2018.meta/train_post_competition.csv')
```

```python
raw_df
```

```python
query_str = 'label != "Knock" and label != "Cough" and label != "Bark" and label != "Meow" and label != "Telephone" and label != "Chime" and label != "Bus"'
```

```python
df = raw_df.query(query_str)
```

```python
df = df.reset_index()
df = df[['fname', 'label', 'manually_verified', 'freesound_id',
       'license']]
df
```

```python
raw_mapping = df['label'].value_counts()
raw_mapping = raw_mapping.index
raw_mapping.to_list()
```

```python
mapping = {v:k for k,v in enumerate(raw_mapping)}
mapping
```

```python
df['classID'] = 0

for i in range(len(df['label'])) :
    df['classID'][i] = mapping[df['label'][i]]

df
```

## file 지우는 함수

```python
def remove_file(f_list, r_list) :
    count = 0
    file_list = os.listdir(f_list)
    
    for file in file_list :
        for remove in r_list :
            if file == remove :
                print(file)
                file_name = f_list+'/'+file
                os.remove(file_name)
                count += 1
            else :
                pass
            
    print('%d' %count, 'has been removed')
```

```python
l_train = df['fname'].to_list()
```

```python
len(l_train)
```

```python
f_list = '/Users/anjaeu/Desktop/UrbanSound8K/FSDKaggle2018.audio_train_erased'
r_list = l_train

remove_file(f_list, r_list)
```
